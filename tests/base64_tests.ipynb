# Base64 encoding and decoding using Python

Resource: [Envato Tuts+ - Base64 Encoding and Decoding Using Python](https://code.tutsplus.com/tutorials/base64-encoding-and-decoding-using-python--cms-25588).

## Required librairies

```python
import base64
from IPython.display import Image, display
```

## Encoding text

```python
text_binary = b'Hey'
print(f'text_binary: {text_binary}')  # b'Hey'

text_b64encode = base64.b64encode(text_binary)
print(f'text_b64encode: {text_b64encode}')  # b'SGV5'
```

## Encoding image

```python
image = open('deer.png', 'rb')
# Open binary file in read mode
image_read = image.read()
image_b64encode = base64.b64encode(image_read)
print(f'image_b64encode: {image_b64encode}')
```

## Decoding image

```python
image_b64decode = base64.b64decode(image_b64encode)
# Create a writable image and write image
image_result = open('deer_decode.png', 'wb') 
# Note that following command returns the size of the image in Bytes
image_result.write(image_b64decode)
```
